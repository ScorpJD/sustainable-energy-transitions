```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np
%matplotlib inline
```

CCGT Gas

```python
fop=np.linspace(12,20,100)
```

```python
fcap=np.linspace(80,140,100)
```

```python
eroi=np.zeros([len(fop),len(fcap)])
```

```python
for i in range(len(fop)):
    for j in range(len(fcap)):
        eroi[i,j]=fop[i]*fcap[j]
```

```python
fig,axes=plt.subplots(2,2,figsize=(13,7))
for i in range(len(axes)):
    for j in range(len(axes[i])):
        ax=axes[i][j]
        im=ax.imshow(eroi[::-1])
        #cax = fig.add_axes([0.27, 0.8, 0.5, 0.02])
        #fig.colorbar(im, cax=cax, orientation='horizontal')
        ax.set_xlabel('fop')
        ax.set_ylabel('fcap')
        ax.set_title('EROEI\n'+str(i))
```
