$f(t)=1/{(1+e^{-\alpha t- \beta})}$

```python
def logist(p,t):
    return [1/(1+np.exp(-p[0]*i-p[1])) for i in np.array(t)-p[2]]
def errfunc(p,t,y):
    return y-logist(p,t)
from scipy import optimize
```

$f(t)=1/{(1+e^{-\alpha \cdot {e^{-\gamma t}} - \beta})}$

```python
def logist2(p,t):
    return [1/(1+np.exp(-p[0]*(np.exp(-p[3]*i))-p[1])) for i in np.array(t)-p[2]]
def errfunc2(p,t,y):
    return y-logist2(p,t)
from scipy import optimize
```

```python
def plotter(ax,x,y,c,l,z=2,zz=2,step=2,w=-50,w2=30,m='o',f=True):
    yrs=range(x[0]-40,x[len(x)-1]+10)
    maxi=[0,0]
    maxv=-100
    
    #try a few initial values for maximum rsquared
    i=0
    for k in range(1,5):
        p0 = [1., 1., x[len(x)*k/5]]
        fit2 = optimize.leastsq(errfunc,p0,args=(x,y),full_output=True)
        ss_err=(fit2[2]['fvec']**2).sum()
        ss_tot=((y-y.mean())**2).sum()
        rsquared=1-(ss_err/ss_tot)
        if rsquared>maxv:
            maxi=[i,k]
            maxv=rsquared

    i=maxi[0]
    k=maxi[1]
    
    p0 = [1., 1., x[len(x)*k/5], -1+i*0.5]
    fit2 = optimize.leastsq(errfunc,p0,args=(x,y),full_output=True)
    ss_err=(fit2[2]['fvec']**2).sum()
    ss_tot=((y-y.mean())**2).sum()
    rsquared=1-(ss_err/ss_tot)

    
    ax.scatter(x[::step],y[::step],lw*3,color=c,marker=m)
    #ax.plot(yrs,logist(fit2[0],yrs),color="#006d2c",lw=lw)
    if f: ax.plot(yrs,logist(fit2[0],yrs),color=c,lw=lw)
    #ax.plot(yrs,logist(fit2[0],yrs),color=c,lw=1)
    
    yk=logist([fit2[0][0],fit2[0][1],fit2[0][2],fit2[0][3]],range(3000))
    mint=0
    maxt=3000
    perc=0.1
    for i in range(3000):
        if yk[i]<perc: mint=i
        if yk[i]<1-perc: maxt=i
    mint=0
    maxt=3000
    perc=0.1
    for i in range(3000):
        if yk[i]<perc: mint=i
        if yk[i]<1-perc: maxt=i
    
    coord=len(x)*zz/5
    if l=='US':g=0.04
    else:
        g=0.35
        if z<-1:
            ax.text(0.04,0.62,'$f(t)=1/{\{1+exp[-\\alpha (t-t_0) - \\beta}]\}$', 
                    horizontalalignment='left',
                    verticalalignment='bottom',
                    transform=ax.transAxes,color="#40386b",fontsize=11) 
    if z<-1:
        ax.text(g,0.6,l+u'\n$R^2 = '+str(np.round(rsquared,2))+'$\n'+\
                    '$\\alpha = '+str(np.round(fit2[0][0],2))+'$\n'+\
                    '$\\beta = '+str(np.round(fit2[0][1],2))+'$\n'+\
                    '', horizontalalignment='left',
                    verticalalignment='top',
                    transform=ax.transAxes,color="#40386b")  
        
    if z<0:
        ax.annotate(l, xy=(yrs[coord], logist(fit2[0],yrs)[coord]),\
                    xycoords='data', 
                    xytext=(w, w2), textcoords='offset points',
                    arrowprops=dict(arrowstyle="->")) 
    yrs=range(x[0]-40,x[len(x)-1]+50)
    return [yrs,logist(fit2[0],yrs)]
```

```python
eq=Eq(Y,(a*F**-r+(1-a)*R**-r)**(-1/r))
eq  #main CES equation
    #Y = value of function - analogous to energy demand?
    #R = renewables
    #F = fossils
    #alfa = fossil share
    #rho = factor related to elasticity of substituion
```

```python
R_sol=solve(eq,R)[0]
R_sol #expression of R from the main CES equation
```

```python
eq2=Eq(s,1/(1+r))
eq2 #definition of elasticity of substituion
    #rho = elasticity of substitution
```

```python
r_sol=solve(eq2,r)[0]
r_sol #exression of rho in terms of sigma
```

```python
[RR_sol,rr_sol]=solve([eq,eq2],[R,r])[0]
RR_sol #expression of R in terms of sigma
```

```python
YR_sol=solve([eq,eq2],[Y,r])[0][0]
YR_sol #expression of Y in terms of sigma
```

```python
def ces(e,refyear,init_F=0.75,s_fix=0.5,zeta=0.0,gamma=0.0):
    years=e[0][refyear-e[0][0]:1991-e[0][0]]
    Y_zero=100 #initial demand
    #F_fix=np.array([init_F*Y_zero*(1.0-1.0/(len(years)-1)*i) for i in range(len(years))])
    #F_fix=init_F*Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    F_fix=Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    Y_fix_zero=YR_sol.subs([[R,Y_zero*(1-init_F)],[F,Y_zero*init_F],[a,init_F],[s,s_fix]])
    Y_fix=np.array([Y_fix_zero*((1+gamma)**i) for i in range(len(years))])
    #new line
    #F_fix=np.array([F_fix[i]*Y_fix[i]/Y_fix_zero for i in range(len(years))])
    #end
    YG=np.array([Y_zero*((1+gamma)**i) for i in range(len(years))])
    R_fix=[RR_sol.subs([[Y,Y_fix[k]],[F,F_fix[k]],[s,s_fix],[a,init_F]]) for k in range(len(F_fix))]
    
    R_fix=[i for i in R_fix if np.complex(i).imag==0]
    R_fix=[i for i in R_fix if i>0]
    
    return [R_fix[:-1],F_fix[:len(R_fix)-1],years[:len(R_fix)-1],YG[:len(R_fix)-1]]
```

```python
def ces2(e,total,refyear,init_F=0.75,s_fix=0.5,zeta=0.0,gamma=0.0):
    years=e[0][refyear-e[0][0]:1991-e[0][0]]
    #Y_zero=100 #initial demand
    Y_zero=total[refyear] #initial demand
    #F_fix=np.array([init_F*Y_zero*(1.0-1.0/(len(years)-1)*i) for i in range(len(years))])
    #F_fix=init_F*Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    F_fix=Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    Y_fix_zero=YR_sol.subs([[R,Y_zero*(1-init_F)],[F,Y_zero*init_F],[a,init_F],[s,s_fix]])
    Y_fix=np.array([Y_fix_zero*((1+gamma)**i) for i in range(len(years))])
    #new line
    F_fix=np.array([F_fix[i]*Y_fix[i]/Y_fix_zero for i in range(len(years))])
    #end
    YG=np.array([Y_zero*((1+gamma)**i) for i in range(len(years))])
    R_fix=[RR_sol.subs([[Y,Y_fix[k]],[F,F_fix[k]],[s,s_fix],[a,init_F]]) for k in range(len(F_fix))]
    
    R_fix=[i for i in R_fix if np.complex(i).imag==0]
    R_fix=[i for i in R_fix if i>0]
    
    return [R_fix[:-1],F_fix[:len(R_fix)-1],years[:len(R_fix)-1],YG[:len(R_fix)-1]]
```

```python
def dces(e,refyear,init_F=0.75,s_fix=0.5,sf=0.01,zeta=0.0,gamma=0.0,step=1):
    #dynamically adjust sigma and resolve equation
    years=e[0][refyear-e[0][0]:1991-e[0][0]]
    Y_zero=100 #initial demand
    #F_fix=np.array([init_F*Y_zero*(1.0-1.0/(len(years)-1)*i) for i in range(len(years))])
    #F_fix=init_F*Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    F_fix=Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    Y_fix_zero=YR_sol.subs([[R,Y_zero*(1-init_F)],[F,Y_zero*init_F],[a,init_F],[s,s_fix]])
    Y_fix=np.array([Y_fix_zero*((1+gamma)**i) for i in range(len(years))])
    #init R_fix, C_fix
    R_fix=[0 for k in range(len(F_fix))]
    C_fix=[0 for k in range(len(F_fix))]
    YG=np.array([Y_zero*((1+gamma)**i) for i in range(len(years))])
    
    s_fix_k=[]
    r_fix_k=[]
    
    for k in range(len(F_fix)):
        t=1
        s_fix_k.append(s_fix) #append to sigma array
        r_fix_k.append(solve(eq2,r)[0].subs(s,s_fix)) #append to rho array
        #update sigma every t years
        if (k%t==0):
            #s_fix*=1+sf #update rho
            s_fix-=sf #update rho
            
    R_fix=[RR_sol.subs([[Y,Y_fix[k]],[F,F_fix[k]],[s,s_fix_k[k]],[a,init_F]]) for k in range(len(F_fix))]
    
    R_fix=[i for i in R_fix if np.complex(i).imag==0]
    R_fix=[i for i in R_fix if i>0]
    
    return [R_fix[:-1],F_fix[:len(R_fix)-1],years[:len(R_fix)-1],YG[:len(R_fix)-1],s_fix_k[:len(R_fix)-1],r_fix_k[:len(R_fix)-1]]
```

```python
eq3=Eq(c_r,c_f/((a/(1-a))*(F/R)**(-1-r)))
eq3 #cost of R compared to F
    #c_f fixed cost of F
```

```python
c_r_sol=solve(eq3,c_r)[0]
c_r_sol #cost of R expressed in terms of rho
```

```python
cR_r_sol=solve([eq3,eq2],[c_r,r])[0][0]
cR_r_sol #cost of R expressed in terms of sigma
```

```python
eq4=Eq(C,c_r*((a/(1-a))*(F/R)**(-1-r))-c_f)
eq4 #carbon tax: relative cost of F compared to R - fixed cost of F
```

```python
C_sol=solve(eq4,C)[0]
C_sol #expression of C in terms of sigma, F, R, a, c_f
```

```python
CR_sol=solve([eq4,eq2],[C,r])[0][0]
CR_sol #expression of C in terms of rho, F, R, a, c_f
```

```python
def dces4(e,refyear=1900,dt=10,init_F=0.95,s_fix=0.5,sf=0.0,zeta=0.0,gamma=0.0,step=1):
    #dynamically adjust sigma and resolve equation
    sfs=s_fix*1.0
    years=range(refyear,refyear+dt)
    Y_zero=100 #initial demand
    #F_fix=np.array([init_F*Y_zero*(1.0-1.0/(len(years)-1)*i) for i in range(len(years))])
    F_fix=init_F*Y_zero*np.array(e[1][refyear-e[0][0]:refyear+dt-e[0][0]])
    #print F_fix,init_F
    #F_fix=Y_zero*np.array(e[1][refyear-e[0][0]:1991-e[0][0]])
    Y_fix_zero=YR_sol.subs([[R,Y_zero*(1-init_F)],[F,Y_zero*init_F],[a,init_F],[s,s_fix]])
    Y_fix=np.array([Y_fix_zero*((1+gamma)**i) for i in range(len(years))])
    #init R_fix, C_fix
    R_fix=[0 for k in range(len(F_fix))]
    C_fix=[0 for k in range(len(F_fix))]
    YG=np.array([Y_zero*((1+gamma)**i) for i in range(len(years))])
    
    s_fix_k=[]
    r_fix_k=[]
    
    for k in range(len(F_fix)):
        t=1
        s_fix_k.append(s_fix) #append to sigma array
        r_fix_k.append(solve(eq2,r)[0].subs(s,s_fix)) #append to rho array
        #update sigma every t years
        s_fix/=1+abs(sf) #update rho
    if sf<0: 
        #print 'k'
        #print s_fix
        s_fix_k=-1.0*np.array(s_fix_k)+2*sfs
            
    R_fix=[RR_sol.subs([[Y,Y_fix[k]],[F,F_fix[k]],[s,s_fix_k[k]],[a,init_F]]) for k in range(len(F_fix))]
    
    R_fix=[i for i in R_fix if np.complex(i).imag==0]
    R_fix=[i for i in R_fix if i>0]
    
    k=0
    c_f_fix=1 #cost of fossil, taken 1 to normalize
    c_r_fix_zero=cR_r_sol.subs([[R,R_fix[k]],[F,F_fix[k]],[s,s_fix_k[k]],[a,init_F],[c_f,c_f_fix]])
    c_r_fix=np.array([c_r_fix_zero*((1-zeta*step)**i) for i in range(len(years))])
    C_fix=[CR_sol.subs([[R,R_fix[k]],[F,F_fix[k]],[s,s_fix_k[k]],[a,init_F],[c_f,c_f_fix],[c_r,c_r_fix[k]]])\
       for k in range(len(R_fix))]
    
    return [R_fix[:-1],F_fix[:len(R_fix)-1],C_fix[:len(R_fix)-1],years[:len(R_fix)-1],YG[:len(R_fix)-1],s_fix_k[:len(R_fix)-1],r_fix_k[:len(R_fix)-1]]
```

```python
gx=[]
gy=[]
gc=[]
x=[0.05]
y=[0.95]
dt=5
for i in range(9):
    [x,y,c,yrs,YY,ys,yr]=dces4(usf,1900+i*dt,dt,y[len(y)-1]*1.0/(x[len(x)-1]+y[len(y)-1]),0.9)
    gx+=list(x)
    gy+=list(y)
    gc+=list(c)
hx=[]
hy=[]
hc=[]
x=[0.05]
y=[0.95]
dt=5
for i in range(9):
    [x,y,c,yrs,YY,ys,yr]=dces4(usf,1900+i*dt,dt,y[len(y)-1]*1.0/(x[len(x)-1]+y[len(y)-1]),0.5)
    hx+=list(x)
    hy+=list(y)
    hc+=list(c)
jx=[]
jy=[]
jc=[]
x=[0.05]
y=[0.95]
dt=5
for i in range(9):
    [x,y,c,yrs,YY,ys,yr]=dces4(usf,1900+i*dt,dt,y[len(y)-1]*1.0/(x[len(x)-1]+y[len(y)-1]),1.9)
    jx+=list(x)
    jy+=list(y)
    jc+=list(c)
```
